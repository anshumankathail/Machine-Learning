```python
import numpy as np
import pandas as pd
```

##### import data

```python
df = pd.read_csv('../data/Wine.csv')

x = df.iloc[:, :-1].to_numpy()
y = df.iloc[:, -1].to_numpy()
```

```python
df.head()
```

##### split data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

##### feature scaling

> scaling is necessary before LDA

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

##### Linear Discriminant Analysis

> supervised learning algorithm

> projects high-dimensional data onto a lower-dimensional subspace that maximizes class separability

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components = 2)

x_train = lda.fit_transform(x_train, y_train)
x_test = lda.transform(x_test)
```

```python
print(x_train)
```

##### train model

```python
from sklearn.svm import SVC

classifier = SVC(kernel = 'rbf')

classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

##### model evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score

print('confusion matrix:\n', confusion_matrix(y_test, y_pred))
print('\naccuracy:', accuracy_score(y_test, y_pred))
print('f1 score:', f1_score(y_test, y_pred, average = 'weighted')) # for multi-class, 'average' argument is a must
```
