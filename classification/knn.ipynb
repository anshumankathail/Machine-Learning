```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### import data

```python
df = pd.read_csv('../data/Social_Network_Ads.csv')

x = df.iloc[:, :-1].to_numpy()
y = df.iloc[:, -1].to_numpy()
```

```python
df.head()
```

##### split data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

##### feature scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

##### train knn classifier

```python
from sklearn.neighbors import KNeighborsClassifier

# for minkowski metric: when p = 1, the distance is manhattan and for p = 2, the distance is euclidean; p -> power parameter
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)

classifier.fit(x_train, y_train)
```

##### predict on new input

```python
print(classifier.predict(sc.transform([[24, 100000]])))
```

```python
print(classifier.predict_proba(sc.transform([[24, 100000]])))
```

##### predict test results

```python
y_pred = classifier.predict(x_test)

np.concatenate((y_test.reshape(-1, 1), y_pred.reshape(-1, 1)), axis = 1)
```

##### confusion matrix, accuracy

confusion matrix:

<img src="../data/images/confusion_matrix.jpg" width="240">

accuracy = (TP + TN) / (TP + TN + FP + FN)

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print("confusion matrix:\n", confusion_matrix(y_test, y_pred))

print("accuracy =", accuracy_score(y_test, y_pred))
```
