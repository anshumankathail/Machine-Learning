```python
import numpy as np
import pandas as pd
```

##### import data

```python
df = pd.read_csv('../data/Cell_Data.csv')

x = df.iloc[:, 1:-1].to_numpy()
y = df.iloc[:, -1].to_numpy()
```

```python
df.head()
```

##### encode labels

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

y = le.fit_transform(y)
```

##### split data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

##### train model

> tree based models do not require feature scaling

```python
from xgboost import XGBClassifier

classifier = XGBClassifier()

classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

##### k-fold cross validation

> bias is error caused because the model is too simple to capture the real pattern

> variance is error caused because the model is too sensitive to training data

```python
from sklearn.model_selection import cross_val_score

accuracies = cross_val_score(estimator = classifier, X = x_train, y = y_train, cv = 10)
```

```python
print(f'validation accuracies: {np.round(accuracies, 2)}')
print(f'\navg accuracy: {accuracies.mean() * 100:.2f} %')
print(f'standard deviation: {accuracies.std() * 100:.2f} %') # can give an idea about variance
```

##### confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score

print('confusion matrix:\n', confusion_matrix(y_test, y_pred))
print(f'\naccuracy = {accuracy_score(y_test, y_pred) * 100 :.2f} %')
print(f'f1 score = {f1_score(y_test, y_pred) :.2f}')
```

> high accuracy on training data = low bias

> high accuracy on training data, low accuracy on avg accuracy accross folds = high variance
