```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### import dataset

```python
df = pd.read_csv('../data/Position_Salaries.csv')
df # data is not big enough to split
```

```python
x = df.iloc[:, 1].to_numpy() # 0-th column is just for understanding
y = df.iloc[:, -1].to_numpy()
```

```python
# convert x and y from 1D to 2D arrays for regressor
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
```

> feature scaling is not needed for decision trees because we are just splitting the 3D space

##### train decision tree regressor model

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state = 0)

regressor.fit(x, y)
```

##### predict on a new input

```python
regressor.predict([[6.5]])
```

##### visualize the model

```python
plt.scatter(x, y, color = 'red')

xn = np.arange(x.min(), x.max(), 0.1).reshape(-1, 1) # for better resolution

plt.plot(xn, regressor.predict(xn))

plt.title('Decision Tree Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()

# since the no. of features is only 1, so the decision tree regressor doesn't fit well in this case
# the model works best for high dimensional data
```
