```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### import the dataset

```python
df = pd.read_csv('../data/Position_Salaries.csv')
df # data isn't big enough to split
```

```python
x = df.iloc[:, 1:-1] # 0-th column of the dataset is just for readability
y = df.iloc[:, -1]
```

##### train the model

train linear regression for comparison

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x, y)
```

train polynomial linear regression

why linear? Because the coefficients in the equation are linear <br>
why polynomial? Because the highest degree of the variables in the equation is greater that 1

```python
# why import from preprocessing? because we are transforming x1 to x1^2, x1^3, ..., x1^n
from sklearn.preprocessing import PolynomialFeatures

pf = PolynomialFeatures(degree = 4) # choose maximum value of n

x_poly = pf.fit_transform(x)

lr2 = LinearRegression()

lr2.fit(x_poly, y)
```

##### visualize linear regression

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lr.predict(x))

plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

##### visualize polynomial linear regression

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lr2.predict(x_poly))

plt.title('Polynomial Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

##### predict on input using linear regression

```python
lr.predict(np.array([[6.5]]))
```

##### predict on input using polynomial linear regression

```python
lr2.predict(np.array(pf.transform([[6.5]])))
```
