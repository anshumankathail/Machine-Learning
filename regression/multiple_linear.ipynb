```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### import dataset

```python
df = pd.read_csv('../data/50_Startups.csv')

x = df.iloc[:, :-1].to_numpy()
y = df.iloc[:, -1].to_numpy()
```

```python
df.head()
```

##### handle categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# no need to use OneHotEncoder(drop='first'), since multiple linear regression library already handles it
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(drop='first'), [3])], remainder='passthrough') 
x = ct.fit_transform(x)
```

```python
x[0]
```

##### split the dataset

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

##### feature scaling not needed since coefficients compensate to put everything on the same scale

##### feature selection not needed as the libraries handle that internally

##### train model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train, y_train)
```

##### predict test results

```python
y_pred = lr.predict(x_test)
```

```python
np.set_printoptions(precision=2)

np.concatenate((y_pred.reshape(-1, 1), y_test.reshape(-1, 1)), axis = 1)
```

##### evaluate model performance

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```

##### make a single prediction

```python
lr.predict(np.array([[1, 0, 150000, 100000, 400000]])) # for state florida
```

##### final linear regression equation

```python
print('b0 =', lr.intercept_)
print('b1, b2, b3, b4, b5, b6 =', lr.coef_)
```
