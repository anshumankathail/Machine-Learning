```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### import dataset

```python
df = pd.read_csv('../Data/Salary_Data.csv')

x = df.iloc[:, :-1].to_numpy()
y = df.iloc[:, -1].to_numpy()
```

```python
df.head()
```

##### split the dataset

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

##### train the model

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(x_train, y_train)
```

##### predict the test results

```python
y_pred = regressor.predict(x_test)
y_pred
```

##### visualize the training set results

```python
plt.title('Salary vs Experience for training set')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')

plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')

plt.show()
```

##### visualize the test set results

```python
plt.title('Salary vs Experience for test set')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')

plt.scatter(x_test, y_test, color = 'red')

# can be replaced with x_test, y_pred as the regression line from the unique equation remains the same
plt.plot(x_train, regressor.predict(x_train), color = 'blue')

plt.show()
```

##### Make a single prediction

```python
# salary of employee with 12 years of experience

regressor.predict(np.array([[12]])) # requires a 2d np array
```

##### Get the values of the coefficients in y = b0 + b1*x

```python
print("b0 = ", regressor.intercept_)
print("b1 = ", regressor.coef_)
```
